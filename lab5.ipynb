\begin{align*}
&\quad \quad 4.2x_1 + 10x_2 = 136 \\
&\quad \quad -10.1x_1 + 13.8x_2 = 132 \\
&\quad \quad 18.3x_1 - 7.6x_2 = 108
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
equations = [
    (4.2, 10, 136, 'yellow', '--'), 
    (-10.1, 13.8, 132, 'purple', '-.'), 
    (18.3, -7.6, 108, 'aqua', ':')
]

fig, axs = plt.subplots(1, 3, figsize=(12, 4))

for i, equation in enumerate(equations, 1):
    a, b, c, color, linestyle = equation
    
    # Вибір підграфіку для поточного графіку
    ax = axs[i-1]
    
    ax.grid(True)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_title(f'Пряма {i}: {a}x1 + {b}x2 = {c}')
    
    x = np.linspace(-10, 10, 100)
    y = (c - a * x) / b
    ax.plot(x, y, color=color, linestyle=linestyle)

plt.tight_layout()
plt.show()
```

```python
equations = [
    (4.2, 10, 136, 'yellow', '--'), 
    (-10.1, 13.8, 132, 'purple', '-.'), 
    (18.3, -7.6, 108, 'aqua', ':')
]

intersection_points = []
for i in range(len(equations)-1):
    for j in range(i+1, len(equations)):
        A = np.array([[equations[i][0], equations[i][1]], [equations[j][0], equations[j][1]]])
        b = np.array([equations[i][2], equations[j][2]])
        intersection_point = np.linalg.solve(A, b)
        intersection_points.append(intersection_point)

plt.figure(figsize=(8, 16), dpi=100)
plt.grid(True, linestyle='--', color='gray')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Перетин прямих')

# Побудова кожної прямої
for equation in equations:
    a, b, c, color, linestyle = equation
    x = np.linspace(-5, 20, 100)
    y = (c - a * x) / b
    plt.plot(x, y, color=color, linestyle=linestyle, label=f'{a}x1 + {b}x2 = {c}')
    if a == 4.2:
        plt.text(x[10], y[5], f'{a}x1 + {b}x2 = {c}', fontsize=8, rotation=-20, rotation_mode='anchor')
    if a == -10.1:
        plt.text(x[10], y[5], f'{a}x1 + {b}x2 = {c}', fontsize=8, rotation=30, rotation_mode='anchor')
    if a == 18.3:
        plt.text(x[10], y[5], f'{a}x1 + {b}x2 = {c}', fontsize=8, rotation=65, rotation_mode='anchor')

# Додавання точок перетину та вказівок
for point in intersection_points:
    plt.plot(point[0], point[1], 'ro')
    plt.annotate(f'({point[0]:.1f}, {point[1]:.1f})', xy=(point[0], point[1]), xytext=(point[0]+1, point[1]+1),
                 arrowprops=dict(arrowstyle='->'))

x = np.linspace(-5, 20, 100)
y1 = (136 - 4.2 * x) / 10
y2 = (132 + 10.1 * x) / 13.8
y3 = (108 - 18.3 * x) / -7.6
y4 = np.maximum(y2, y3)
plt.fill_between(x, y3, y4, color='gray', alpha=0.5)
y5 = np.maximum(y1, y3)
plt.fill_between(x, y3, y5, color='white')

plt.legend()
plt.savefig('intersection_plot.jpg')
plt.savefig('intersection_plot.png')
plt.savefig('intersection_plot.svg')
plt.show()
```
